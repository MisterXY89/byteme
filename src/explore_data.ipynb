```python
from recommend import Recommender
import matplotlib.pyplot as plt
import numpy as np
import bayespy.plot as bpplt
import pandas as pd
```

```python
df = pd.read_csv("/Users/anton/Documents/Projects/byteme/data/dummy.csv").iloc[:,:2]
```

```python
model = Recommender(sim=True)
model.fit()
```

```python
res1 = model.model["methods"]["candidate group assignment probs"]
res2 = model.model["research"]["candidate group assignment probs"]

a = res1._message_to_child()[0].reshape(35,6)
b = res2._message_to_child()[0].reshape(35,6)
```

```python
import matplotlib.pyplot as plt
import numpy as np
person = 5

with plt.style.context("seaborn-whitegrid"):
    fig, ax = plt.subplots(1,2, constrained_layout = True, figsize =(10,3))
    ax[0].scatter(np.arange(0,a.shape[1],1), a[person,:], label = "Method Group")
    ax[0].scatter(np.arange(0,b.shape[1],1), b[person,:], label = "Research Group")
    ax[1].scatter(np.arange(0,weighted.shape[1],1), weighted[person,:], color="green")
    ax[0].legend()
    ax[0].set_title("Comparison of Cluster Output for both topics")
    ax[0].set_xticks(np.arange(0,a.shape[1],1))
    ax[0].set_xticklabels(["Cluster 1", "Cluster 2", "Cluster 3", "Cluster 4", "Cluster 5", "Cluster 6"])
    ax[1].set_title("Comparison of weighted Cluster Output")
    ax[1].set_xticks(np.arange(0,a.shape[1],1))
    ax[1].set_xticklabels(["Cluster 1", "Cluster 2", "Cluster 3", "Cluster 4", "Cluster 5", "Cluster 6"])
    fig.supylabel("Probability of belonging to Cluster")
    plt.show()
```

```python
weighted = a.copy()
for person in range(a.shape[0]):
    weighted[person,:] = a[person,:] * df.iloc[person, 0]+ b[person,:] * 1-df.iloc[person, 0]
```
