```python
from recommend import Recommender
import matplotlib.pyplot as plt
import numpy as np
import bayespy.plot as bpplt
import pandas as pd
```

```python
model = Recommender(sim=True)
result = model.fit(cluster_size=5)
```

```python
import random 

method = result["method"]
research = result["research"]

with plt.style.context("seaborn-whitegrid"):
    fig, ax = plt.subplots(1,2, constrained_layout = True, figsize =(10,3))

    for xe, ye in zip(np.arange(0,method.shape[1],1), method.T):
        x = np.linspace(xe-.2,xe-.05,len(ye))
        ax[0].scatter(x, ye, c="gray")
    for xe, ye in zip(np.arange(0,research.shape[1],1), research.T):
        x = np.linspace(xe+.05,xe+.2,len(ye))
        ax[0].scatter(x, ye, c="orange")
    plt.show()
```

```python
method = result["method"]
research = result["research"]
weighted = result["weighted"]


person = 27



with plt.style.context("seaborn-whitegrid"):
    fig, ax = plt.subplots(1,2, constrained_layout = True, figsize =(10,3))
    ax[0].scatter(np.arange(0,method.shape[1],1), method[person,:], label = "Method Group")
    ax[0].scatter(np.arange(0,method.shape[1],1), research[person,:], label = "Research Group")
    ax[1].scatter(np.arange(0,weighted.shape[1],1), weighted[person,:], color="green")
    ax[0].legend()
    ax[0].set_title("Comparison of Cluster Output for both topics")
    
    ticks = np.arange(0,method.shape[1],1)
    labels = [f"Cluster {x+1}" for x in range(method.shape[1])]
    
    ax[0].set_xticks(ticks)
    ax[1].set_xticks(ticks)
    ax[0].set_xticklabels(labels)
    ax[1].set_xticklabels(labels)
    ax[1].set_title("Comparison of weighted Cluster Output")
    fig.supylabel("Probability of belonging to Cluster")
    plt.show()
```
